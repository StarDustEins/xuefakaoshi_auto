```python
soup = bs4.BeautifulSoup(session.get('http://xxpt.scxfks.com/study/courses/all').text, 'lxml')
```

```python
ss = soup.find('div', class_='sectoral subject')
ss
```

```python
for i in ss.descendants:
    print(type(i), i)
```

```python
dz = {}
for i in ss.descendants:
    # print(i.string)
    # print(i.name)
    if i.name == 'a':
        dz.update({i.string: i.attrs['name']})
dz
```

```python
table = soup.find('tbody')
table
```

```python
ll = []
for tr in table.findAll('tr'):
    for a in tr.findAll('a'):
        if 'href' in a.attrs:
            print(a.attrs)
            ll.append(a.attrs['href'])
        else:
            ll.append('-')
```

```python
df = pandas.read_html(session.get('http://xxpt.scxfks.com/study/courses/all').text)[0]
df
```

```python
df['type'] = df['分类'].map(dz)
df['link'] = ll
df
```

```python
sub_soup = bs4.BeautifulSoup(session.get('http://xxpt.scxfks.com' + '/study/course/3718').text)
```

```python
perday = 0
for i, row in df.iterrows():
    # print(i, row['link'], row['type'])
    if row['link'] != '-':
        for uu in sub_soup.find('ul', id='chapter-none').find_all('a'):
            print(uu.string, uu.attrs['href'])
            get_res = session.get('http://xxpt.scxfks.com' + uu.attrs['href'])
            chapter_index = uu.attrs['href'].rfind('/') + 1
            post_aaa_res = session.post(
                'http://xxpt.scxfks.com/study/learnlog/%s/%s' % (uu.attrs['href'][chapter_index:], row['type']))
            print(uu.attrs['href'][chapter_index:], post_aaa_res.text)
            if perday > 10:
                break
            print('---------------------------------------------------------------' + '/n')
            perday += 1
            print(perday)
```
